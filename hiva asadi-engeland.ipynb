```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("EnglandWeather.csv")
data
```

```python
df=pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
df2=pd.DataFrame(data, columns= ['Summary','Precip Type'])
df2
```

```python
df2.describe()
```

```python
df3=df.dropna(how='any')
df4=df2.dropna(how='any')
df3
```

```python
df4
```

```python
plt.figure(figsize=(6,6), dpi=90)
plt.scatter(df3['Summary'],df3['Humidity'])
plt.xlabel("Summary")
plt.xticks(rotation=90)
plt.ylabel("Humidity")
plt.grid()
plt.show()

```

```python
plt.figure(figsize=(6,6), dpi=90)
plt.scatter(df3['Precip Type'],df3['Humidity'])
plt.xlabel("Precip Type")
plt.xticks(rotation=90)
plt.ylabel("Humidity")
plt.grid()
plt.show()

```

```python
Year= pd.to_datetime(df3.Date).dt.year
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
plt.title('Humidity of year')
sns.countplot(x=Year)
```

```python
Mon= pd.to_datetime(df3.Date).dt.month
```

```python
plt.figure(figsize=(10,8))
plt.title('Humidity of month')
sns.countplot(x=Mon)
```

```python
dd= pd.to_datetime(df3.Date).dt.day
```

```python
plt.figure(figsize=(10,8))
plt.title('Humidity of day')
sns.countplot(x=dd)
```
