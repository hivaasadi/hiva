```python
#به صورت تمريني يه ساختار سه بعدي هم تمرين کردم و در حد توانم نوشتم هر قسمت دستور چه کارهايي ميکنه
#البته در هر نوع نمودار  سه  بعدي تنظيمات مختلفي داره من روي اين نمودار توضيحات رو نوشتم
```

```python

import matplotlib.pyplot as plt
import numpy as np

# رفع حالت تصادفي براي تکرار پذيري
np.random.seed(10000)

#تنظيمات ظاهري سيستم رو مشخص ميکنيم اگر پيش فرض نخواستيم

fig = plt.figure(figsize=[4,6] , dpi=150 , facecolor='blue', edgecolor='green')

#ساختار رو مشخص ميکنه که سه بعدي هستش که در يک متغير ذخيره ميشه و در آخر دستورات براي نمايش ازش استفاده ميشه

ax = fig.add_subplot(projection='3d') 
#رنج انتخاب ايکس و ايگرگ رو نشون ميده اون 4 آخر هم تعداد ستون نمايش داده رو نشون ميده مثلا به جاي عدد 4 بنويسيم 1 فقط 1 ستون نمايش ميده

x, y = np.random.rand(2,10) * 4


#ساختار ستون و سطرها رو مشخص ميکنيم  مثلا ماتريس [4*4] 
# (bins)تعداد ستون هاي بلند شده رو مشخص ميکنه

hist, xedges, yedges = np.histogram2d(x, y, bins=4 ,range=[[0, 4], [0, 4]])

#ساختار شبکه مش رو مشخص ميکنيم مثلا لبه هاي ايکس و ايگرگ و ساخت نمايه که همون آ و جي آخري هستش 
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")

# تابع راول آرايه هاي موجود را بر ميگرداند مانند اين مثال
# x = np.array([[1, 2, 3], [4, 5, 6]])
# np.ravel(x)
# array([1, 2, 3, 4, 5, 6])

xpos = xpos.ravel()  
ypos = ypos.ravel()
zpos = 0

#اينجا هم ميگه که نمودار زد رو به سمت بالا به صورت نيم  به بالا نشون بده مثلا اول بنويس صفر بعد نيم بعد يک بعد يک و نيم و... البته اينم براي اينه که ايکس و ايگرگ رو در زد ضرب کرده 
dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.ravel()

#اينجا هم چيزاي رو که ميخاد ميگه برام ترسيم کن به صورت سه بعدي

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')

plt.show()
```
